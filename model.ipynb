```python
# 🚀 Install required libraries
!pip install transformers torch matplotlib seaborn numpy -q

# 📌 Import necessary modules
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import pipeline

# 🎯 Load Pretrained Sentiment Analysis Models
roberta_model = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment-latest")
bertweet_model = pipeline("sentiment-analysis", model="finiteautomata/bertweet-base-sentiment-analysis")
distilbert_model = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# 🔍 Function to Generate Synthetic Social Media Data
def generate_synthetic_data():
    return [
        "I absolutely love this product! It's amazing. 🌟",
        "Worst experience ever. I will never buy this again. 😡",
        "The new update is decent, but I liked the old version better.",
        "Amazing service and fast delivery! Highly recommend. 🚀🔥",
        "The product is okay, nothing special. Neutral experience.",
        "Terrible customer support. No help at all! Avoid at all costs. 😠",
        "Super happy with my purchase. Would buy again. 😍",
        "Not sure how I feel about this product. It's just okay.",
    ]

# 📌 Fetch Synthetic Social Media Texts
texts = generate_synthetic_data()

# 🔍 Perform Sentiment Analysis Using All Three Models
results_roberta = [roberta_model(text)[0] for text in texts]
results_bertweet = [bertweet_model(text)[0] for text in texts]
results_distilbert = [distilbert_model(text)[0] for text in texts]

# 🔥 Optimized Sentiment Selection Algorithm (Weighted Confidence)
final_results = []
confidence_threshold = 0.60  # Threshold for Neutral classification

for text, r, b, d in zip(texts, results_roberta, results_bertweet, results_distilbert):
    labels = [r['label'], b['label'], d['label']]
    scores = [r['score'], b['score'], d['score']]

    # Majority Voting
    label_counts = {label: labels.count(label) for label in set(labels)}
    most_common_label = max(label_counts, key=label_counts.get)

    # Confidence-Based Selection
    if label_counts[most_common_label] > 1:
        final_label = most_common_label
        final_score = max(scores)
    else:
        max_index = scores.index(max(scores))
        final_label = labels[max_index]
        final_score = scores[max_index]

    # Handling Uncertain Cases
    if final_score < confidence_threshold:
        final_label = "NEUTRAL"
        final_score = 0.70

    final_results.append({"label": final_label, "score": final_score})

# 🎨 Visualizing Sentiment Distribution (Bar Chart)
sentiments = [res['label'] for res in final_results]
scores = [res['score'] for res in final_results]

plt.figure(figsize=(8, 5))
sns.barplot(x=sentiments, y=scores, palette="coolwarm")
plt.xlabel("Sentiment")
plt.ylabel("Confidence Score")
plt.title("📊 Optimized Sentiment Analysis (Ensemble Learning)")
plt.show()

# 📊 Pie Chart for Sentiment Ratio
sentiment_counts = {s: sentiments.count(s) for s in set(sentiments)}
plt.figure(figsize=(6, 6))
plt.pie(sentiment_counts.values(), labels=sentiment_counts.keys(), autopct='%1.1f%%', colors=['green', 'red', 'gray'])
plt.title("📊 Sentiment Distribution")
plt.show()

# ✅ Print Optimized Sentiment Results
for text, res in zip(texts, final_results):
    print(f"\n📝 Text: {text}\n🔍 Sentiment: {res['label']} (Confidence: {res['score']:.2f})")
```
